```python
def set_time(time):
    list =[]
    for i in range(3):
        list.append(time//(60**(2-i)))
        time = time%(60**(2-i))
    list.append(time)
    return " : ".join(map(str, list))
def get_time(obj: PI, matriz: Matriz_base):
    start = time.time()
    obj.start(matriz)
    end = time.time()
    
    return set_time(end - start)
```
